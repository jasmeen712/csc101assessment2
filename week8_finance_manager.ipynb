```python
def load_transactions(filename):
    """Read lines from a transaction file safely."""
    try:
        # Open the file in read mode
        with open(filename, 'r') as file:
            lines = file.readlines() # Read all lines from the file
        # Strip whitespace from each line and filter out empty lines
        return [line.strip() for line in lines if line.strip()]
    except FileNotFoundError:
        # Handle the case where the file does not exist
        print(f"File '{filename}' not found!")
        return [] # Return an empty list if file not found
```

```python
def parse_transaction_line(line):
    """Parses a single transaction line into a dictionary.
    Handles lines like 'Item,-Amount' and returns {'item': 'Item', 'amount': -Amount}.
    Returns (dictionary, None) for success, (None, error_message) for corrupted lines.
    """
    try:
        # Split the line into item and amount by comma
        parts = line.split(',')
        # Check if the line has exactly two parts
        if len(parts) != 2:
            raise ValueError("Line does not contain exactly two parts separated by a comma.")

        item = parts[0].strip() # Get item name and remove leading/trailing whitespace
        amount = float(parts[1].strip()) # Convert amount to a float and remove whitespace

        return {'item': item, 'amount': amount}, None # Return parsed transaction and no error
    except (ValueError, IndexError) as e:
        # Catch errors during splitting or type conversion
        return None, f"Error: {e}" # Return None for transaction and the error message

def parse_transaction_lines(lines):
    """Parses a list of transaction lines into a list of dictionaries.
    Returns a list of parsed transactions and a list of (original_line, error_message) for failures.
    """
    parsed_transactions = []
    failed_details = []
    for line in lines:
        transaction_result, error_message = parse_transaction_line(line) # Attempt to parse each line
        if transaction_result:
            parsed_transactions.append(transaction_result) # Add successful parses to the list
        else:
            failed_details.append((line, error_message)) # Add original line and error message to failures
    return parsed_transactions, failed_details

# Example usage:
transaction_lines = [
    'Coffee,-4.50',
    'Groceries,-25.75',
    'Salary,1500.00',
    'Invalid line without comma',
    'Rent,-1200',
    'Bonus,500.00',
    'Corrupted value,abc'
]

# Parse the example transaction lines
parsed_data, failed_transactions = parse_transaction_lines(transaction_lines)
print("Parsed Transactions:")
for t in parsed_data:
    print(t)

print(f"\nNumber of failed lines: {len(failed_transactions)}")
if failed_transactions:
    print("Details of failed lines:")
    for line, error_msg in failed_transactions:
        print(f"  - Original line: '{line}' | {error_msg}")
```

```python
# Week 8: Smart Personal Finance Manager

def load_transactions(filename):
    """Read lines from a transaction file safely."""
    try:
        # Open the file in read mode
        with open(filename, 'r') as file:
            lines = file.readlines() # Read all lines from the file
        # Strip whitespace from each line and filter out empty lines
        return [line.strip() for line in lines if line.strip()]
    except FileNotFoundError:
        # Handle the case where the file does not exist
        print(f"File '{filename}' not found!")
        return [] # Return an empty list if file not found

def parse_transaction_line(line):
    """Parses a single transaction line into a dictionary.
    Handles lines like 'Item,-Amount' and returns {'item': 'Item', 'amount': -Amount}.
    Returns (dictionary, None) for success, (None, error_message) for corrupted lines.
    """
    try:
        # Split the line into item and amount by comma
        parts = line.split(',')
        # Check if the line has exactly two parts
        if len(parts) != 2:
            raise ValueError("Line does not contain exactly two parts separated by a comma.")

        item = parts[0].strip() # Get item name and remove leading/trailing whitespace
        amount = float(parts[1].strip()) # Convert amount to a float and remove whitespace

        return {'item': item, 'amount': amount}, None # Return parsed transaction and no error
    except (ValueError, IndexError) as e:
        # Catch errors during splitting or type conversion
        return None, f"Error: {e}" # Return None for transaction and the error message

def parse_transaction_lines(lines):
    """Parses a list of transaction lines into a list of dictionaries.
    Returns a list of parsed transactions and a list of (original_line, error_message) for failures.
    """
    parsed_transactions = []
    failed_details = []
    for line in lines:
        transaction_result, error_message = parse_transaction_line(line) # Attempt to parse each line
        if transaction_result:
            parsed_transactions.append(transaction_result) # Add successful parses to the list
        else:
            failed_details.append((line, error_message)) # Add original line and error message to failures
    return parsed_transactions, failed_details

def analyze_transactions(transactions):
    """Analyzes a list of transaction dictionaries to calculate totals.
    Returns total number of transactions, total income, and total expenses.
    """
    total_transactions = len(transactions) # Count the total number of valid transactions
    total_income = 0.0 # Initialize total income
    total_expenses = 0.0 # Initialize total expenses

    # Iterate through each transaction to categorize as income or expense
    for transaction in transactions:
        amount = transaction['amount']
        if amount > 0:
            total_income += amount # Add positive amounts to income
        else:
            total_expenses += amount # Add negative amounts (or zero) to expenses

    return total_transactions, total_income, total_expenses

def process_transactions(lines):
    """Process transaction lines into clean data and report failures."""
    parsed_data, failed_transactions = parse_transaction_lines(lines)

    # Report failed lines to the user
    if failed_transactions:
        print(f"\nWarning: {len(failed_transactions)} lines could not be processed.")
        print("Details of failed lines:")
        for line, error_msg in failed_transactions:
            print(f"  - Original line: '{line}' | {error_msg}")

    return parsed_data

def show_summary(transactions):
    """Display transaction analysis results to the user in a professional format."""
    if not transactions:
        print("No valid transactions to summarize.")
        return

    # Analyze the successfully parsed transactions
    total_trans, income, expenses = analyze_transactions(transactions)
    net_amount = income + expenses # Calculate net amount

    # Display the summary professionally
    print("\n--- Transaction Summary ---")
    print(f"Number of transactions processed: {total_trans}")
    print(f"Total Income:               {income:10.2f}") # Formatted to 2 decimal places, right-aligned
    print(f"Total Expenses:             {expenses:10.2f}") # Formatted to 2 decimal places, right-aligned
    print(f"---------------------------")
    print(f"Net Amount:                 {net_amount:10.2f}") # Formatted to 2 decimal places, right-aligned
    print("---------------------------")

# Main program entry point
def main():
    print("üè¶ Personal Finance Manager")
    print("=" * 30)

    # Define the filename for transactions
    filename = "transactions.txt"
    # Load raw lines from the file
    raw_lines = load_transactions(filename)

    if not raw_lines:
        print("No data to process or file not found!")
        return

    # Process the raw data, handling corrupted lines
    transactions = process_transactions(raw_lines)

    # Show the summary of valid transactions
    show_summary(transactions)

# --- Example Usage for parse_transaction_lines ---
# This section demonstrates how to use the parse_transaction_lines function
# independently with a hardcoded list of transaction strings.

print("\n--- Demonstrating parse_transaction_lines ---")
example_transaction_lines = [
    'Coffee,-4.50',
    'Groceries,-25.75',
    'Salary,1500.00',
    'Invalid line without comma',
    'Rent,-1200',
    'Bonus,500.00',
    'Corrupted value,abc',
    'Another Item,10.00'
]

# Parse the example transaction lines
parsed_example_data, failed_example_transactions = parse_transaction_lines(example_transaction_lines)

print("\nParsed Transactions from example_transaction_lines:")
for t in parsed_example_data:
    print(t)

print(f"\nNumber of failed lines in example: {len(failed_example_transactions)}")
if failed_example_transactions:
    print("Details of failed lines in example:")
    for line, error_msg in failed_example_transactions:
        print(f"  - Original line: '{line}' | {error_msg}")


# Run the main program when the script is executed
if __name__ == "__main__":
    main()
```

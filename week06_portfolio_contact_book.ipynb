```python
#B1
Contacts = {
    "John": {"phone": "0401 765 865", "email": "john@gmail.com"},
     "Jacob": {"phone": "0867 765 754", "email": "jacob@gmail.com"},
     "Tom": {"phone": "0456 765 983", "email": "tom@gmail.com"},
}
#B2
for name, contact_info in Contacts.items():
    print(f"{name}: {contact_info}")
#B3
name_to_lookup = input("Enter the name to look up: ")

if name_to_lookup in Contacts:
    contact_info = Contacts[name_to_lookup]
    print(f"Phone: {contact_info['phone']}")
    print(f"Email: {contact_info['email']}")
else:
    print(f"Contact '{name_to_lookup}' not found.")
```
